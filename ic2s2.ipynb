```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
in_df = pd.read_csv('California_in_region_percentage_by_week.csv')
out_df = pd.read_csv('California_out_region_percentage_by_week.csv')
```

```python
in_df
```

```python
out_df = out_df.set_index('TractID')
```

```python
dates = [datetime.strptime(i, '%Y-%m-%d') for i in in_df.columns.values]
```

```python
dates
```

```python
for row in in_df.iloc:
    plt.figure(figsize=(25, 25), dpi = 500)
    values = [float(i[:-1]) for i in row.values]
    plt.plot(dates, values)
    plt.xticks(dates, rotation=30, horizontalalignment='right')
    plt.title(row.name)
    plt.axvline(x = datetime(2021, 12, 27), color = 'r', label = 'axvline - full height')
    plt.axvline(x = datetime(2022, 1, 1), color = 'r', label = 'axvline - full height')
    plt.axvline(x = datetime(2021, 9, 29), color = 'r', label = 'axvline - full height')
    plt.axvline(x = datetime(2021, 9, 24), color = 'r', label = 'axvline - full height')
    plt.axvline(x = datetime(2022, 1, 7), color = 'r', label = 'axvline - full height')
    plt.savefig(f'{row.name}.jpg')
    plt.close()
    
```
